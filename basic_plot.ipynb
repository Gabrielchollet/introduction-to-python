```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)

plt.show()
```

`pyplot` is a sub-package of `matplotlib`, hence the dot. 

```python
# Customization
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
# Add more data
year = [1800, 1850, 1900] + year
pop = [1.0, 1.262, 1.650] + pop

plt.plot(year, pop)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0, 2, 4, 6, 8], ['0', '2B', '4B', '6B', '8B'])

plt.show()
```

# Scatter Plot:

The resulting scatter plot simply plots all the individual data points. Hence Python doesn't connect the dots with a line. 

```python
plt.scatter(year, pop)
plt.show()
```

# Histogram:

The histogram is a type of visualization that is very useful to explore your data and get an idea about the distribuition of your variables. 
Futhermore the number of bins is pretty important. Too few bins will oversimplify reality and won't show you the details. Too many bins will overcomplicate reality and won't show the bigger picture.

```python
values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.hist(values, bins=3)
plt.show()
```
